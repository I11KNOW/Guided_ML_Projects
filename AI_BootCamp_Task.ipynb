#### Good job uptil now! remember, in order to fit our model to the data we should split it into training set and test set for evaluation.

```python
# split your dataset into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1)
```

#### 4th: The model (Linear Regression)

```python
# make an object for your linear regression model and fit it to the training data (features, target)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# make predictions using the test data features
y_predicted = model.predict(x_test)
```

```python
plt.scatter(x_test , y_test, color='blue')
plt.plot(x_test, y_predicted, color='red')
plt.title('Data points Vs Predicted Line')
plt.xlabel('Features')
plt.ylabel('Labels (Targets)')
```

```python
# Time to evaluate our model using the rmse metric (it's okay to use only mse without the root as well)
from sklearn.metrics import mean_squared_error
from math import sqrt
M=mean_squared_error(y_test , y_predicted)
R=sqrt(M)
R
```
